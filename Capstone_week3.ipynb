# Part1

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
wiki = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(wiki,'lxml')
print(soup.title)
from IPython.display import display_html
df = str(soup.table)
display_html(df,raw=True)
```

```python
canada_data = pd.read_html(df)
canada_data = canada_data[0]
```

```python
canada_data
```

```python
canada_data=canada_data[canada_data["Borough"] != 'Not assigned']
```

```python
canada_data
canada_data = canada_data.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
canada_data.reset_index(inplace=True)
```

```python
canada_data['Neighborhood'] = np.where(canada_data['Neighborhood'] == 'Not assigned',canada_data['Borough'], canada_data['Neighborhood'])
```

```python
canada_data.shape
```

# Part 2

```python
geo_data= pd.read_csv('https://cocl.us/Geospatial_data')
```

```python
data = pd.merge(canada_data, geo_data, on = 'Postal Code')
data.head()
```

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighborhood in zip(t_data['Latitude'],t_data['Longitude'],t_data['Borough'],t_data['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
```

```python
k=5
t_cluster = t_data.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(t_cluster)
kmeans.labels_
t_data.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
t_data
```

```python
# create map
clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighborhood, cluster in zip(t_data['Latitude'], t_data['Longitude'], t_data['Neighborhood'], t_data['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(clusters)
```
